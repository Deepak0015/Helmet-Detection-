```python
from ultralytics import YOLO
import cv2 
```

```python
model =  YOLO('/home/saaho/THYTECH/HelmetDetection/HelmetDetection/Model/weights2/best.pt')
```

```python
video_path =  '/home/saaho/THYTECH/HelmetDetection/HelmetDetection/istockphoto-488557289-640_adpp_is.mp4'
cap =  cv2.VideoCapture(video_path)
```

```python
import cv2
import torch
from ultralytics import YOLO


device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("End of video stream.")
        break

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    results = model(rgb_frame, conf=0.25)  

    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0]) 
            conf = box.conf[0].item()  
            class_id = int(box.cls[0]) 
            class_name = model.names[class_id]

            color = (0, 255, 0) if class_name == "helmet" else (0, 0, 255)
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, f"{class_name}: {conf:.2f}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    cv2.imshow("Helmet Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Get video writer to save output (optional)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('helmet_detection_output.avi', fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform YOLOv8 detection
    results = model(frame)[0]

    # Draw bounding boxes
    for box in results.boxes.data:
        x1, y1, x2, y2, conf, cls = map(int, box[:6])  # Extract bounding box values
        if conf > 0.5:  # Confidence threshold
            label = f"{model.names[cls]} ({conf:.2f})"
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Show output
    cv2.imshow("Helmet Detection", frame)
    out.write(frame)  # Save video output

    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to exit
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2
import torch
from ultralytics import YOLO

model = YOLO("/home/saaho/THYTECH/HelmetDetection/HelmetDetection/Model/weights2/best.pt")

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read frame.")
        break

    results = model(frame , imgsz =320) 
  
    annotated_frame = results[0].plot(line_width=2)

    cv2.imshow("Webcam Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

```
